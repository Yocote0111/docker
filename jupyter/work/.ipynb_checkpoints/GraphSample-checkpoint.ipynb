```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
# %matplotlib inline
%matplotlib notebook
```

```python
def func(x):
    return 1 - (1-1/x)**(1/x)
    
x = np.arange(1, 100, 1)
y = func(x)

plt.plot(x,y)
```

```python
x = np.arange(-5,5,0.1)
y = np.arange(0,20,0.1)
x,y = np.meshgrid(x,y)
z = x + y

# カラーメッシュグラフ (x,y,z=color)
fig = plt.figure()
ax = fig.gca()
im = ax.pcolormesh(x, y, z)                # draw: z = f(x,y)
fig.colorbar(im)
```

```python
x = np.arange(-5,5,0.1)
y = np.arange(0,20,0.1)
x,y = np.meshgrid(x,y)
z = x**2 + y

# 3Dグラフ (x, y, z)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(y,x,z, cmap=plt.cm.viridis) # draw: z = f(x,y)

# カラーメッシュグラフ (x,y,z=color)
fig = plt.figure()
ax = fig.gca()
im = ax.pcolormesh(x, y, z)                # draw: z = f(x,y)
fig.colorbar(im)
```

```python
a = 0.07
b = 1
c1 = 0.02
c2 = 0.01
p1 = 8
p2 = 17
q1 = 0
q2 = 0.3

def fx(x):
    return a*x * b

def fy(y):
    f1 = c1*(y-p1)**2 + q1
    f2 = c2*(y-p2)**2 + q2
    return np.where( f1 >= f2, f2, f1)

def f(x,y):
    return fx(x)+fy(y)
```

```python
x = np.arange(0, 30, 0.5)

# 折れ線グラフ
fig = plt.figure()
ax = fig.gca()
ax.plot(x, fx(x))

fig = plt.figure()
ax = fig.gca()
ax.plot(x, fy(x))
```
