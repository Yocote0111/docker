```python
x = np.linspace(-3, 3, 50)


fx = stats.norm.pdf(x) # pdf (probability density function): 確率密度関数 f(x)
Fx = stats.norm.cdf(x) # cdf (cumulative distribution function): 累積分布関数 F(x)
Sx = stats.norm.sf(x) # sf (survival function): 生存関数 S(x) = 1 - F(x)

log_fx = stats.norm.logpdf(x) # log(f(x))
log_Fx = stats.norm.logcdf(x) # log(F(x))
log_Sx = stats.norm.logsf(x)  # log(S(x))

# 可視化
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, fx, label=r'$ f(x) $')
ax1.plot(x, Fx, label=r'$ F(x) $')
ax1.plot(x, Sx, label=r'$ S(x) $')
ax1.grid(True)
ax1.legend()
ax2.plot(x, log_fx, label=r'$ log(f(x)) $')
ax2.plot(x, log_Fx, label=r'$ log(F(x)) $')
ax2.plot(x, log_Sx, label=r'$ log(S(x)) $')
ax2.grid(True)
ax2.legend()
plt.show()
```

# 任意の確率分布のfitting

```python
c = (3/4)**(2/3)
def func(x):
    return -x**2+c

def i_func(x):
    return -x**3/3+c*x

def my_dist_pdf(x):
    y = func(x)
    return np.where( y<0, 0, y)

def my_dist_cdf(x):
    y = i_func(x)-i_func(-c)
    y = np.where( x<-c, 0, y)
    y = np.where( x>c, 1, y)
#    y = np.where( y<0, 0, y)
#    y = np.where( y>1, 1, y)
    return y

def my_dist_trunc_pdf(x, a, b):
    scale = my_dist_cdf(b)-my_dist_cdf(a)
    y = my_dist_pdf(x)/scale
    y = np.where( x<a, 0, y)
    y = np.where( x>b, 0, y)
    return y

def my_dist_trunc_cdf(x, a, b):
    scale = my_dist_cdf(b)-my_dist_cdf(a)
    m = i_func(a) if a > -c else i_func(-c)
    y = (i_func(x)-m)/scale
    y = np.where( x<-c, 0, y)
    y = np.where( x>c, 1, y)
    y = np.where( y<0, 0, y)
    y = np.where( y>1, 1, y)
    return y
```

```python
x = np.linspace(-1.5, 1.5, 100)
fx = my_dist_pdf(x)
Fx = my_dist_cdf(x)
a = -0.4
b = 0.6
ftx = my_dist_trunc_pdf(x, a, b)
Ftx = my_dist_trunc_cdf(x, a, b)

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, fx, label='pdf')
ax1.plot(x, ftx, label='pdf_t')
ax1.grid(True)
ax1.legend()
ax2.plot(x, Fx, label='cdf')
ax2.plot(x, Ftx, label='cdf_t')
ax2.grid(True)
ax2.legend()
plt.show()
```

## 以下をオーバーライドする

### デフォルト定義
rv_generic
 _ppf(q) :=  F( _cdf(x))    # 自動計算 ( cdf からの探索によるもの )
 _isf(q) := _ppf(1.0-q)     # 自動定義
 _logcdf(x) = log( _cdf(x)) # 自動定義
 _sf(x) :=  1.0 - _cdf(x)   # 自動定義
 _logsf(x):= log( _sf(x))   # 自動定義

rv_continuous
 _cdf(x) :=  F( _pdf(x))    # 自動計算 ( pdf の積分によるもの )
 _pdf(x) :=  F( _cdf(x))    # 自動定義 ( pdf の微分によるもの )
 _logpdf(x):= log( _pdf(x)) # 自動定義

## fit の内容
rv_continuous
 fit 
 - 最尤推定量(MLE: Maximize Likelihood Estimate) によるもの
  - 罰則付き負の対数尤度  -sum(logpdf(x)) + penalty の最小化で pdf さえあれば算出可能
 - 指定したパラメータを固定(freeze)できる (floc=1, fscale=1, fa=0, ..)
 - optimizer を指定できる (optimizer=myfunc) 
  -  デフォルトは optimize.fmin でシンプレクス法によるもの
 - パラメータの初期値を指定できる (
 
 




```python
from scipy.stats._distn_infrastructure import rv_continuous

class my_dist(rv_continuous):
    def _pdf(self, x, p1, p2):
        return my_dist_trunc_pdf(x, p1, p2)
    
    def _cdf(self, x, p1, p2):
        return my_dist_trunc_cdf(x, p1, p2)
    
    def _argcheck(self, *args):
        return args[0] < args[1]
```

```python
dist = my_dist(name='my dist')
s = dist.rvs(size=100, scale=1, loc=0, p1=-0.4, p2=0.6) # 確率分布にしたがってランダムサンプリング

# 可視化
fig, ax1 = plt.subplots(1, 1)
ax1.hist(s, histtype='stepfilled', alpha=0.2)
plt.show()
```
